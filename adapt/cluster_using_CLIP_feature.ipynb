```python
import os
import pickle
import numpy as np

from sklearn import preprocessing
from sklearn.cluster import KMeans
```

```python
category_path = "../clip_feature"
```

```python
img_name = []
img_feature = []
class_name = []

for folder in os.listdir(category_path):
    folder_path = os.path.join(category_path, folder)
    for file in os.listdir(folder_path):
        img_name.append(file)
        img_feature.append(np.load(os.path.join(folder_path, file)))
        class_name.append(folder)
```

```python
len(img_name)
```

```python
img_feature = np.array(img_feature).squeeze(1)
```

```python
img_feature.shape
```

```python
# using img_feature_norm instead of img_feature have effect like using cosine distance instead of eucildean distance in KMeans.
num_cluster = 125

img_feature_norm = preprocessing.normalize(img_feature)

cluster = KMeans(n_clusters=num_cluster)
cluster.fit(img_feature_norm)
```

```python
cluster.cluster_centers_.shape
```

```python
with open('../storage/cluster.pkl','wb') as f:
    pickle.dump(cluster, f)
```

```python
np.save('../storage/cluster_centers_.npy', cluster.cluster_centers_)
```

```python
import pandas as pd
```

```python
res = cluster.predict(img_feature_norm)
```

```python
res
```

```python
tmp_dict = {
    'image_name': img_name,
    'cluster_idx': res,
    'class': class_name
}

df = pd.DataFrame(data=tmp_dict)
df
```
