```python
import os
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('../storage/cluster.csv')
df
```

```python
df.columns
```

```python
for folder in os.listdir('../features_npy'):
  if len(os.listdir(os.path.join('../features_npy', folder))) != 100:
    print(folder)
```

```python
num_cluster = 125
centroid_embed = []

for cluster_idx in range(num_cluster):
  print(cluster_idx)
  index_per_class = df.index[df['cluster_idx']==cluster_idx]
  image_embed_per_cluster = []
  for idx in index_per_class:

    image_embed_per_cluster.append(np.load(f"../features_npy/{df['class'][idx]}/{df['image_name'][idx]}"))

  print(len(image_embed_per_cluster))
  centroid_embed.append(np.mean(image_embed_per_cluster, axis=0))
```

```python
tmp = np.array(centroid_embed)
```

```python
tmp.shape
```

```python
np.save('../storage/centroid_embed.npy', tmp)
```
