```python
def retrieval(query = None, gallery = 'TUBerlin', model = None):
  if query is None:
    raise ValueError("Please provide query image (path)!")

  query = Image.open(query)
  query_img = preprocess(query, sk)
  query_img = query_img

  scores = []
  if gallery == 'TUBerlin':
   gallery_path = os.path.join("/content/ZSE-SBIR/datasets", "TUBerlin", "ImageResized_ready")
   category_list = os.listdir(gallery_path)

   for folder in category_list:
      img_path = os.path.join(gallery_path, folder)
      img_list = os.list_dir(img_path)

      for image in img_list:
          img = preprocess(os.path.join(img_path, image), im)

          _ ,score = model(query.to(device), img.to(device))

          scores.append([score, os.path.join(img_path, image)])
```
